```python
#4. Seleccionar todo de la tabla "customers" donde "city" sea "London" o "Seattle". Almacenar los resultados en una variable.

mycursor = cnx.cursor()
query_3 ="""SELECT * FROM customers WHERE country = "UK" OR country = "USA" """ # city = "London" OR city = "Seattle" 
mycursor.execute(query_3)
results_3 = mycursor.fetchall()

column_names = [col[0] for col in mycursor.description] #forma rara que hemos encontrado para nombrar las columnas


db_results_3 = pd.DataFrame(results_3,columns=column_names)
db_results_3
```

```python
#tail sacara las últimas filas 
db_results_3.tail(10)
```

```python
#sample saca registros al azar
db_results_3.sample(5)
```

```python
# Hemos usado un método isnull para resolver si hay nulls en la tabla

mycursor = cnx.cursor()
query_5 ="""SELECT * FROM customers WHERE region IS NULL OR fax IS NULL """ 
mycursor.execute(query_5)
results_5 = mycursor.fetchall()

column_names_1 = [col[0] for col in mycursor.description]


db_results_5 = pd.DataFrame(results_5,columns=column_names_1)
db_results_5.isnull()
```
